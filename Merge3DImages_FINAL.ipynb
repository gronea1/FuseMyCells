```python
import tifffile
import numpy as np
import matplotlib.pyplot as plt
from skimage import measure
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

!pip install "napari[all]"
import napari
```

```python
# Load the 3D fused image
fused_image = tifffile.imread("/Users/maryannjohnson/Library/CloudStorage/Box-Box/Documents/Fuse My Cells Challenge/image_169-180/image_171_membrane_angle.tif")

# Compute the maximum intensity projection along the z-axis
mip_image = np.max(fused_image, axis=0)

# Save the 2D image as PNG
output_path = "/Users/maryannjohnson/Library/CloudStorage/Box-Box/Documents/Fuse My Cells Challenge/image_169-180/image_171_membrane_angle_mip.png"
plt.imsave(output_path, mip_image, cmap='gray')

print(f"Saved MIP image to {output_path}")

# Display the projection using matplotlib
plt.imshow(mip_image, cmap='gray')
plt.title('Maximum Intensity Projection')
plt.axis('off')
plt.show()
```
